- 1. What is the overall survival rate of passengers aboard the Titanic?

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("https://raw.githubusercontent.com/springboardmentor943x/ImpactSense-Intern-project/refs/heads/main/Milestone_1/Week_2/Day_6/titanic.csv")

# Calculate overall survival rate
survival_rate = df['Survived'].mean() * 100

print(f"Overall survival rate: {survival_rate:.2f}%")
```

- 3. What is the average age of passengers who survived vs. those who did not survive?

```python
import pandas as pd

# Load dataset (update path or use raw GitHub link)
df = pd.read_csv("https://raw.githubusercontent.com/springboardmentor943x/ImpactSense-Intern-project/refs/heads/main/Milestone_1/Week_2/Day_6/titanic.csv")

# Calculate average age for survivors (1) and non-survivors (0)
avg_age = df.groupby("Survived")["Age"].mean()

print("Average Age of Passengers:")
print("Did not survive (0):", round(avg_age[0], 2))
print("Survived (1):", round(avg_age[1], 2))
```

- 4. How many passengers embarked from each port (Embarked), and what are their survival rates?

```python
import pandas as pd

# Load dataset (update path or use raw GitHub link)
df = pd.read_csv("https://raw.githubusercontent.com/springboardmentor943x/ImpactSense-Intern-project/refs/heads/main/Milestone_1/Week_2/Day_6/titanic.csv")

# Number of passengers from each embarkation port
passenger_counts = df["Embarked"].value_counts()

# Survival rates by embarkation port
survival_rates = df.groupby("Embarked")["Survived"].mean() * 100

# Combine both into one table
result = pd.DataFrame({
    "Passenger_Count": passenger_counts,
    "Survival_Rate (%)": survival_rates.round(2)
})

print(result)
```

- 5. What is the distribution of fare prices paid by passengers, and does fare impact survival chances?

```python
import pandas as pd
import matplotlib.pyplot as plt


# Load dataset (update path or GitHub raw link)
df = pd.read_csv("https://raw.githubusercontent.com/springboardmentor943x/ImpactSense-Intern-project/refs/heads/main/Milestone_1/Week_2/Day_6/titanic.csv")

# --- 1. Distribution of Fare Prices ---
plt.figure(figsize=(8,5))
plt.hist(df["Fare"], bins=50, edgecolor="black")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.title("Distribution of Fare Prices")
plt.show()

# --- 2. Compare Fare by Survival ---
plt.figure(figsize=(8,5))
df.boxplot(column="Fare", by="Survived")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Fare")
plt.title("Fare vs Survival")
plt.suptitle("")  # remove extra title
plt.show()

# --- 3. Average Fare by Survival ---
avg_fare = df.groupby("Survived")["Fare"].mean()
print("Average Fare Paid:")
print("Did not survive (0):", round(avg_fare[0], 2))
print("Survived (1):", round(avg_fare[1], 2))
```

- 8. Show the age distribution of passengers using a histogram or KDE plot.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/springboardmentor943x/ImpactSense-Intern-project/refs/heads/main/Milestone_1/Week_2/Day_6/titanic.csv")

# Drop missing ages if any
ages = df['Age'].dropna()

plt.figure(figsize=(8,5))

# Histogram
sns.histplot(ages, bins=30, kde=False, color="skyblue", edgecolor="black")
plt.title("Age Distribution of Titanic Passengers (Histogram)")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()

# KDE Plot
plt.figure(figsize=(8,5))
sns.kdeplot(ages, shade=True, color="purple")
plt.title("Age Distribution of Titanic Passengers (KDE Plot)")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show()
```

- 10. Visualize survival count by embarkation port (Embarked) with a stacked bar chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/springboardmentor943x/ImpactSense-Intern-project/refs/heads/main/Milestone_1/Week_2/Day_6/titanic.csv")

# Create a crosstab of Embarked vs Survived
embarked_survival = pd.crosstab(df['Embarked'], df['Survived'])

# Plot stacked bar chart
embarked_survival.plot(kind='bar', stacked=True, figsize=(8,5), color=["tomato", "skyblue"])

plt.title("Survival Count by Embarkation Port")
plt.xlabel("Embarkation Port")
plt.ylabel("Number of Passengers")
plt.legend(["Did Not Survive", "Survived"])
plt.show()
```

- 11. Display a heatmap of the correlation matrix of numerical features in the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("https://raw.githubusercontent.com/springboardmentor943x/ImpactSense-Intern-project/refs/heads/main/Milestone_1/Week_2/Day_6/titanic.csv")

# Select only numerical columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
